The scikit-learn API requires:
+ firstly that you have the data as a NumPy array or pandas DataFrame. 
+ It also requires that the features take on continuous values, such as the price of a house, as opposed to categories, such as 'male' or 'female'. 
+ It also requires that there are no missing values in the data
+ The target needs to be a single column with the same number of observations as the feature data.

```python
# Import package
from urllib.request import urlretrieve
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data"
```

```python
# Save file locally

urlretrieve(url, "data_voting.csv")
# Read file into a DataFrame and print its head
df = pd.read_csv('data_voting.csv', sep=',')
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns = ['party', 'infants', 'water', 'budget', 'physician', 'salvador',
       'religious', 'satellite', 'aid', 'missile', 'immigration', 'synfuels',
       'education', 'superfund', 'crime', 'duty_free_exports', 'eaa_rsa']
```

```python
df.head()
```

```python
plt.figure()
sns.countplot(x='education', hue='party', data=df, palette='RdBu')
plt.xticks([0,1], ['No', 'Yes'])
plt.show()
```
